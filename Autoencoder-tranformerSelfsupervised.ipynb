```python
import os
import glob
import numpy as np
import librosa
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, davies_bouldin_score, confusion_matrix, ConfusionMatrixDisplay, classification_report
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense, Dropout, LayerNormalization, MultiHeadAttention
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.regularizers import l2
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.keras.optimizers import Adam
import pandas as pd
import seaborn as sns
# Train a supervised classifier using the derived cluster features
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from scipy.signal import butter, filtfilt

# Label mapping
label_mapping = {
    0: ("Normal", "Not Stressed"),
    1: ("NoQueen", "Stressed"),
    2: ("Swarm", "Stressed"),
}
def bandpass_filter(y, sr, lowcut=200, highcut=5000):
    nyquist = 0.5 * sr
    low = lowcut / nyquist
    high = highcut / nyquist
    b, a = butter(4, [low, high], btype='band')
    return filtfilt(b, a, y)

# Function to extract audio features
def extract_audio_features(file_path):
    try:
        y, sr = librosa.load(file_path, sr=None)
        if sr < 16000:
                y = librosa.resample(y, orig_sr=sr, target_sr=16000)
                sr = 16000

        # Apply band-pass filter
        y = bandpass_filter(y, sr, lowcut=200, highcut=5000)

        # MFCCs
        mfcc = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=40)
        mfcc = np.mean(mfcc.T, axis=0)  # Shape: (40,)

        # Delta and Delta-Delta MFCCs
        delta_mfcc = librosa.feature.delta(mfcc)
        delta2_mfcc = librosa.feature.delta(mfcc, order=2)

        # RMS
        rms = librosa.feature.rms(y=y).mean()

        # Spectral Centroid
        spectral_centroid = librosa.feature.spectral_centroid(y=y, sr=sr).mean()

        # Dominant Frequency (STFT)
        stft = np.abs(librosa.stft(y))
        frequencies = librosa.fft_frequencies(sr=sr)
        dominant_frequency = frequencies[np.argmax(np.sum(stft, axis=1))]
        print(f"File: {file_path}")
        print(f"  Dominant Frequency: {dominant_frequency}")
        print(f"  Spectral Centroid: {spectral_centroid}")
        print(f"  RMS: {rms}\n")
        stressVal = "Non-stressed"

        # Combined condition for Stressed classification
        if 350 <= dominant_frequency <= 500 or 950 <= spectral_centroid <= 1200:
            stressVal = "Stressed"

        # Combined condition for Non-stressed classification
        elif 200 <= dominant_frequency <= 300 or 650 <= spectral_centroid <= 900:
            stressVal = "Non-stressed"


        print(f"Stress Value Assigned for each file: {stressVal}")

        # Combine all features into a 1D array
        return np.concatenate([mfcc, delta_mfcc, delta2_mfcc, [rms, spectral_centroid, dominant_frequency]])
    except Exception as e:
        print(f"Error processing {file_path}: {e}")
        return None

# Process audio files
def process_audio_files(audio_dir):
    features = []
    file_names = []
    for file_path in glob.glob(f"{audio_dir}/**/*.*", recursive=True):
        if file_path.lower().endswith(('.wav', '.mp3')):
            feature_vector = extract_audio_features(file_path)
            if feature_vector is not None:
                features.append(feature_vector)
                file_names.append(os.path.basename(file_path))
    return np.array(features), file_names

# Transformer Encoder Block
def transformer_encoder_block(inputs, num_heads, ff_dim, dropout_rate=0.3):
    attention_output = MultiHeadAttention(num_heads=num_heads, key_dim=inputs.shape[-1])(inputs, inputs)
    attention_output = Dropout(dropout_rate)(attention_output)
    attention_output = LayerNormalization(epsilon=1e-6)(inputs + attention_output)

    ffn_output = Dense(ff_dim, activation="relu", kernel_regularizer=l2(0.02))(attention_output)
    ffn_output = Dense(inputs.shape[-1], kernel_regularizer=l2(0.02))(ffn_output)
    ffn_output = Dropout(dropout_rate)(ffn_output)
    return LayerNormalization(epsilon=1e-6)(attention_output + ffn_output)

# Build Transformer-based Autoencoder
def build_transformer_autoencoder(input_dim, encoding_dim, num_heads=4, ff_dim=64, dropout_rate=0.3):
    inputs = Input(shape=(input_dim,))
    x = Dense(encoding_dim, activation='relu', kernel_regularizer=l2(0.02))(inputs)
    x = Dropout(dropout_rate)(x)
    x = tf.keras.layers.Lambda(lambda x: tf.expand_dims(x, axis=1))(x)

    for _ in range(3):
        x = transformer_encoder_block(x, num_heads=num_heads, ff_dim=ff_dim, dropout_rate=dropout_rate)

    x = tf.keras.layers.Lambda(lambda x: tf.squeeze(x, axis=1))(x)
    encoded = Dense(encoding_dim, activation='relu', kernel_regularizer=l2(0.02))(x)
    encoded = Dropout(dropout_rate)(encoded)

    decoded = Dense(input_dim, activation='sigmoid')(encoded)

    autoencoder = Model(inputs, decoded)
    encoder = Model(inputs, encoded)
    autoencoder.compile(optimizer=Adam(learning_rate=0.0005), loss='mse')
    return autoencoder, encoder

# Fine-tune Transformer Autoencoder
def fine_tune_transformer_autoencoder(autoencoder, features_scaled):
    print("Fine-tuning Transformer-based Autoencoder...")
    early_stopping = EarlyStopping(monitor='loss', patience=5, restore_best_weights=True)
    autoencoder.fit(features_scaled, features_scaled, validation_split=0.2, epochs=30, batch_size=32, shuffle=True, verbose=1, callbacks=[early_stopping])

# Plot Training and Validation Metrics
def plot_training_metrics(train_losses, val_losses):
    plt.figure(figsize=(10, 5))
    plt.plot(train_losses, label='Training Loss', color='blue')
    plt.plot(val_losses, label='Validation Loss', color='red')
    plt.xlabel("Epochs")
    plt.ylabel("Loss")
    plt.title("Training and Validation Loss over Epochs")
    plt.legend()
    plt.grid(True)
    plt.show()

# Plot Confusion Matrix
def plot_confusion_matrix_final(all_labels, all_preds, label_mapping):
    cm = confusion_matrix(all_labels, all_preds)
    labels = [v[0] for v in label_mapping.values()]

    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=labels, yticklabels=labels)
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.title('Confusion Matrix for Final Evaluation')
    plt.show()

# Plot Classification Report
# Plot Classification Report
def plot_classification_report(all_labels, all_preds, label_mapping):
    unique_labels = np.unique(all_labels)
    # Filter label_mapping to only include relevant labels
    filtered_label_mapping = {k: v for k, v in label_mapping.items() if v[0] in unique_labels}

    # If filtered_label_mapping is empty, use unique_labels
    if not filtered_label_mapping:
        target_names = unique_labels.tolist()
    else:
        target_names = [v[0] for v in filtered_label_mapping.values()]

    report = classification_report(
        all_labels,
        all_preds,
        target_names=target_names,  # Use filtered target names
        output_dict=True
    )
    classes = list(report.keys())[:-3]
    precision = [report[cls]['precision'] for cls in classes]
    recall = [report[cls]['recall'] for cls in classes]
    f1_score = [report[cls]['f1-score'] for cls in classes]

    bar_width = 0.25
    x = np.arange(len(classes))

    plt.figure(figsize=(12, 6))
    plt.bar(x - bar_width, precision, bar_width, label='Precision', color='blue')
    plt.bar(x, recall, bar_width, label='Recall', color='green')
    plt.bar(x + bar_width, f1_score, bar_width, label='F1-Score', color='orange')
    plt.xticks(x, classes, rotation=45)
    plt.ylabel('Score')
    plt.title('Precision, Recall, and F1-Score per Class')
    plt.legend()
    plt.grid(axis='y')
    plt.show()

# Generate Classification Report as a Table
def generate_classification_report_table(all_labels, all_preds, label_mapping):
    # Get unique classes in all_labels
    unique_labels = sorted(list(set(all_labels)))

    # Dynamically create target names based on unique labels
    target_names = [label_mapping[label][0] for label in range(len(label_mapping)) if label in unique_labels]

    # Generate classification report
    report = classification_report(
        all_labels,
        all_preds,
        labels=unique_labels,  # Pass the actual unique labels
        target_names=target_names,  # Match target names to the unique labels
        output_dict=True
    )

    # Convert the report into a pandas DataFrame
    report_df = pd.DataFrame(report).transpose()
    report_df = report_df[['precision', 'recall', 'f1-score', 'support']]  # Select relevant columns

    # Display the DataFrame
    print("\nClassification Report as a Table:")
    print(report_df)

    # Optionally, save it to a CSV file
    report_csv = "/content/drive/MyDrive/Colab Notebooks/piping_database/classification_report.csv"
    report_df.to_csv(report_csv)
    print(f"Classification Report saved to {report_csv}")

# Define criteria for determining cluster labels using multiple features
def determine_cluster_labels(cluster_features):
    dominant_frequency = cluster_features["average_dominant_frequency"]
    spectral_centroid = cluster_features["average_spectral_centroid"]
    reconstruction_error = cluster_features["average_reconstruction_error"]

    # Debugging: Print values
    print(f"dominant_frequency: {dominant_frequency}")
    print(f"spectral_centroid: {spectral_centroid}")
    print(f"reconstruction_error: {reconstruction_error}")

    # Initialize stress value
    stressVal = "Non-stressed"
    var="NV"

    # Combined condition for Stressed classification
    if 350 <= dominant_frequency <= 500 or 950 <= spectral_centroid <= 1200:
       var="DF"
       stressVal = "Stressed"

    # Combined condition for Non-stressed classification
    elif 200 <= dominant_frequency <= 300 or 650 <= spectral_centroid <= 900:
        var="DF2"
        stressVal = "Non-stressed"

    # Use reconstruction error as a tie-breaker
    elif reconstruction_error > 0.02:
        var="RE"
        stressVal = "Stressed"

    print(f"Variable used for stress classification: {var}")
    print(f"Stress Value Assigned id: {stressVal}")
    return stressVal


# Main Workflow
def main():
    audio_dir = "/content/drive/MyDrive/Colab Notebooks/piping_database/FinalDataSet/"
    print("Processing audio files...")
    features, file_names = process_audio_files(audio_dir)

    if features.size > 0:
        scaler = MinMaxScaler()
        features_scaled = scaler.fit_transform(features)

        input_dim = features_scaled.shape[1]
        encoding_dim = 32
        autoencoder, encoder = build_transformer_autoencoder(input_dim, encoding_dim)

        print("Training Transformer-based Autoencoder...")
        early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
        history = autoencoder.fit(features_scaled, features_scaled, validation_split=0.2, epochs=50, batch_size=64, shuffle=True, verbose=1, callbacks=[early_stopping])

        # Fine-tune the model
        fine_tune_transformer_autoencoder(autoencoder, features_scaled)

        latent_features = encoder.predict(features_scaled)
        reconstruction = autoencoder.predict(features_scaled)

        reconstruction_error = np.mean(np.square(features_scaled - reconstruction), axis=1)
        threshold = np.percentile(reconstruction_error, 95)

        anomaly_labels = ['Stressed' if error > threshold else 'Not Stressed' for error in reconstruction_error]

        kmeans = KMeans(n_clusters=2, random_state=42)
        cluster_labels = kmeans.fit_predict(latent_features)

        # Extract spectral features (example indices for illustration)
        spectral_centroid = features[:, -2]  # Example: Spectral Centroid
        dominant_frequency = features[:, -1]  # Example: Dominant Frequency
        print("Original Dominant Frequencies:", dominant_frequency)
        print("Original Spectral Centroids:", spectral_centroid)
        print(f"Dominant Frequency - Min: {dominant_frequency.min()}, Max: {dominant_frequency.max()}, Mean: {dominant_frequency.mean()}")
        print(f"Spectral Centroid - Min: {spectral_centroid.min()}, Max: {spectral_centroid.max()}, Mean: {spectral_centroid.mean()}")

        # Calculate average spectral features for each cluster
        cluster_spectral_features = {}
        for cluster_id in np.unique(cluster_labels):
            cluster_indices = np.where(cluster_labels == cluster_id)[0]
            cluster_spectral_features[cluster_id] = {
                "average_dominant_frequency": np.max([dominant_frequency[i] for i in cluster_indices]),
                "average_spectral_centroid": np.mean([spectral_centroid[i] for i in cluster_indices]),
                "average_reconstruction_error": np.mean([reconstruction_error[i] for i in cluster_indices]),  # Optional
            }



        # Combine features into a dataset
        X = np.array([[f["average_dominant_frequency"], f["average_spectral_centroid"], f["average_reconstruction_error"]] for f in cluster_spectral_features.values()])
        y = ["Stressed" if "Stressed" in determine_cluster_labels(f) else "Non-stressed" for f in cluster_spectral_features.values()]

        # Split data
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

        # Train Random Forest Classifier
        clf = RandomForestClassifier(random_state=42)
        clf.fit(X_train, y_train)

        # Evaluate
        print("Classifier Accuracy:", clf.score(X_test, y_test))

        # Optional: Use the classifier to predict the stress state for new data
        y_pred = clf.predict(X_test)

        # Map the results back to cluster labels (if needed)
        cluster_labels_mapped = [clf.predict([[f["average_dominant_frequency"], f["average_spectral_centroid"], f["average_reconstruction_error"]]])[0] for f in cluster_spectral_features.values()]

        # Map clusters to labels based on frequency criteria
        cluster_mapping = {}
        for cluster_id, features in cluster_spectral_features.items():
            cluster_mapping[cluster_id] = determine_cluster_labels(features)

        # Apply the mapping to cluster labels
        cluster_labels_mapped = [cluster_mapping[label] for label in cluster_labels]

        print("Cluster Spectral Features (Original Values):")
        for cluster_id, features in cluster_spectral_features.items():
            print(f"Cluster {cluster_id}: {features}")


        true_labels = cluster_labels_mapped  # Simulated true labels

        plot_confusion_matrix_final(true_labels, cluster_labels_mapped, label_mapping)

        plot_training_metrics(history.history['loss'], history.history['val_loss'])

        silhouette_avg = silhouette_score(latent_features, cluster_labels)
        davies_bouldin_avg = davies_bouldin_score(latent_features, cluster_labels)

        print(f"Silhouette Score: {silhouette_avg:.4f}")
        print(f"Davies-Bouldin Index: {davies_bouldin_avg:.4f}")

        generate_classification_report_table(true_labels, cluster_labels_mapped, label_mapping)
        # Visualize Dominant Frequency
        sns.histplot(dominant_frequency, kde=True, bins=20)
        plt.title("Dominant Frequency Distribution")
        plt.xlabel("Frequency (Hz)")
        plt.show()

        # Visualize Spectral Centroid
        sns.histplot(spectral_centroid, kde=True, bins=20)
        plt.title("Spectral Centroid Distribution")
        plt.xlabel("Frequency (Hz)")
        plt.show()

        # Visualize Reconstruction Error
        sns.histplot(reconstruction_error, kde=True, bins=20)
        plt.title("Reconstruction Error Distribution")
        plt.xlabel("Reconstruction Error")
        plt.show()

        results_df = pd.DataFrame({
            'File Name': file_names,
            'Cluster': cluster_labels,
            'Reconstruction Error': reconstruction_error,
            'Stress State': cluster_labels_mapped,
            'Anomaly Label': anomaly_labels
        })

        output_csv = "/content/drive/MyDrive/Colab Notebooks/piping_database/bee_stress_analysis_with_fine_tuning.csv"
        results_df.to_csv(output_csv, index=False)
        print(f"Results saved to {output_csv}")

if __name__ == "__main__":
    main()
```
